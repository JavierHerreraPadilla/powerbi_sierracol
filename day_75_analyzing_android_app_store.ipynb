```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
"""Cargar el dataframe"""

df = pd.read_csv('apps.csv')
```

```python
"""Checkeo inicial de datos"""

print(f'hay {df["Category"].nunique()} tipos de categorías de aplicaciones')
df['Category'].value_counts()
```

```python
"""Checkeo inicial de datos"""

df.sample(5) #ver 5 filas al azar
```

```python
#elimianar las columnas 'Last_Updated' y 'Android_Ver'. No son relevantes apra este análisis

print(df.columns)
df.drop(['Last_Updated', 'Android_Ver'], axis=1, inplace=True
        )
```

```python
df.head()
```

```python
"""Checkeo inicial de datos- cuántas filas tienen datos faltantes en ratings"""
df.Rating.isna().sum()
```

```python
"""limpiar datos - apps que no tienen rating"""

df_clean = df.dropna()
df_clean.shape
```

```python
#"""Verificar datos duplicados""" check for duplicates

print(df_clean.duplicated(), end='\n\n')
print(f'duplicados: {df_clean.duplicated().sum()}')
```

```python
#"""Verificar datos duplicados por el nombre de la app""" check for duplicates

filter = df_clean.duplicated(subset='App', keep=False)
# keep= "first" o "last" o False. "first" marks duplicates as True except for the first occurrence, "last" marks duplicates as True except for the last occurrence, False marks all duplicates as True
df_clean[filter].value_counts()
```

```python
#""" Organizar el df basado en ratings y revews """sorting the df based on ratings and the on reviews

df_clean.sort_values(by=['Rating', 'Reviews'], ascending=False)
```

```python
#"""App de mayor tamaño """largest app size
df_clean.Size_MBs.max()
```

```python
#""" app con la mayor cantidad de reviews """ apps with highest number of reviews
df_clean.sort_values(by='Reviews', ascending=False).head(50)
```

```python
# """ ver si en las primeras 50 ordenadas por el mayor numero de reseñas hay alguna que sea paga""" checks for purchasable apps in top 50
filter2 = df_clean.sort_values(by='Reviews', ascending=False).head(50)['Type'] == 'Free'
df_clean.sort_values(by='Reviews', ascending=False).head(50)[~filter2] #filter2 busca las "free". Para buscar las pagas se niega el filtro
#No hay pagas en las primeras 50 por numero de reseñas
```

```python
import plotly.express as px
```

```python
ratings = df_clean['Content_Rating'].value_counts()
fig = px.pie(labels=ratings.index, values=ratings.values, title="Numero de apps por categoría /App qtty per category", names=ratings.index, hole=0.5) #hole es para crear una dona, hole=0 crea un piechart relleno
fig.update_traces(textposition="inside", textinfo="percent+label")
fig.show()
```

```python
#""" Tipo de datos contenido en la columna Installs """datatype of install column: String
df_clean.Installs
```

```python
#""" conteo de número de app por cada nivel de intalación """count number of apps at each level of instalation
df_clean.Installs.value_counts()
```

```python
#""" quitar las comas de la coluimna installs """remove commas from Installs and count
df_clean.Installs = df_clean.Installs.astype(str).str.replace(',','')
df_clean.Installs = pd.to_numeric(df_clean.Installs)
df_clean[['App', 'Installs']].groupby('Installs').count()
```

```python
#""" convertir price a tipo numérico"""converting the price to numeric data
df_clean.Price = df_clean.Price.astype(str).str.replace('$','')
df_clean.Price = pd.to_numeric(df_clean.Price)
df_clean.info()
```

```python
#""" top de apps más costosas """top most expensive apps
df_clean.sort_values(by='Price', ascending=False).head(20)
```

```python
# """borrar duplicados  """ delete duplicates
df_clean.drop_duplicates(subset='App', keep='first', inplace=True)
print(df_clean.shape)
df_clean[df_clean['App']=="Instagram"]
```

```python
#""" quitar precios mayores a 250 """ removing prices greater than 250
df_clean = df_clean[df_clean.Price < 250]
df_clean.sort_values(by='Price', ascending=False).head(20)
```

```python
# Crear un nueva columna para las ganancias generadas por la app - crate new col to estimate app revenue
df_clean['Revenue_estimate'] = df_clean.Installs.mul(df_clean.Price)
df_clean.sort_values(by='Revenue_estimate', ascending=False)
```

```python
# categorías diferentes - number of different categories
print(f'number of categories: {df_clean.Category.nunique()}')
top10_category = df_clean.Category.value_counts()[:10]
top10_category
```

```python
#Visualización de los datos en gráfico de barras - visualize data in bar chart
bar = px.bar(x=top10_category.index, y=top10_category.values)
bar.show()
```

```python
# agrupamos por categoria de apps y sumamos las instalaciones de cada categoria usando groupby y agg
category_installs = df_clean.groupby('Category').agg({'Installs':pd.Series.sum})
category_installs.sort_values(by='Installs', ascending=True, inplace=True)
category_installs
```

```python
# GRAFICAMOS LOS DATOS ANTERIORES
bar2 = px.bar(x=category_installs.Installs, y=category_installs.index, orientation='h')
bar2.update_layout(xaxis_title='number of downloads', yaxis_title='Category')
bar2.show()
```

```python
#Combinamos los 2 dataframes - merging 2 data frames

cat_number = df_clean.groupby('Category').agg({'App':pd.Series.count})
cat_number_merged = pd.merge(cat_number, category_installs, on='Category', how='inner')
cat_number_merged.sort_values('Installs', ascending=False)
cat_number_merged
```

```python
# Creamos una gráfica de dispersión - creating scatter plot
scatter = px.scatter(cat_number_merged, x='App', y='Installs', title='Concentración por categorías/Category Concentration', size='App',
                     hover_name=cat_number_merged.index, color='Installs')
scatter.update_layout(xaxis_title='Num of Apps (Lower= more concentrated)',yaxis_title='Installs', yaxis=dict(type='log'))
```

```python
#extraer columnas anidadas  - extracting nested columns data using .stack()
print(f' there are {df_clean.Genres.nunique()} genres')
print(f' value counts: {df_clean.Genres.value_counts().sort_values(ascending=True)}')
```

```python
#Dvidimos el texto por el punto y coma - split the string on the semi-colon and then .stack them
stack = df_clean.Genres.str.split(';', expand=True).stack()
print(f'we now have a single column with shape: {stack.shape}')
num_genres = stack.value_counts()
print(f'num of genres {len(num_genres)}')
num_genres
```

```python
#Gráficos por género

bar3 = px.bar(x = num_genres.index[:15],
              y = num_genres.values[:15],
              title='Top Genres',
              hover_name=num_genres.index[:15],
              color = num_genres.values[:15],
              color_continuous_scale='Agsunset')
bar3.update_layout(xaxis_title='Genre',
                   yaxis_title='Number of apps',
                   coloraxis_showscale=False)

bar3.show()
```

```python
 # agrupamos y contamos por columna Type - gruoped bar charts and box plots
 df_clean.Type.value_counts()
```

```python
#agrupamos por category y luego por type y luego contamos - group by category and type and count
df_free_vs_paid = df_clean.groupby(['Category', 'Type'], as_index=False).agg({'App':pd.Series.count})
df_free_vs_paid.sort_values('App')
```

```python
#GRAFICAMOS APPS PAGAS Y GRATIS POR CATEGORÍA
bar4 = px.bar(df_free_vs_paid,
              x='Category',
              y='App',
              title='Free vs Paid Apps by category',
              color='Type',
              barmode='group')
bar4.update_layout(xaxis={'categoryorder':'total descending'},
                   yaxis=dict(type='log'))
bar4.show()
```

```python
# Box plots para la cantidad de instalaciones -  box plots for the number of installs
box = px.box(df_clean,
             y='Installs',
             x='Type',
             color='Type',
             notched=True,
             points='all',
             title='Diferencia de instalaciones por tipo Gratis/paga - How many downloads are paid apps giving up?')
box.update_layout(yaxis=dict(type='log'))
box.show()
```

```python
#Ganancias de app pagas por categoría / How much can paid apps earn?
box2 = px.box(df_clean[df_clean['Type']=='Paid'],
             y='Revenue_estimate',
             x='Category',
             color='Category',
             notched=False,
             title='Ganancias de app pagas por categoría / How much can paid apps earn?')
box2.update_layout(xaxis={'categoryorder': 'min ascending'},
    yaxis=dict(type='log'))
box2.show()
```

```python
#Precios de apps por categoría - app pricing by category
df_paid_app = df_clean[df_clean['Type']=='Paid']
df_paid_app.Price.median()
```

```python
#Precio de las apps por categoría - app price per category
box3 = px.box(df_paid_app,
              x='Category',
              y='Price',
              title='Precio por categoría / price per category',)

box3.update_layout(xaxis={'categoryorder':'max descending'},
                   yaxis=dict(type='log'))
box3.show()
```
